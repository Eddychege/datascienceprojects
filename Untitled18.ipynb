```python
import pandas as pd

# Assuming df2 is your "not shared" DataFrame

# Convert 'Date (Created)' to datetime objects (handle potential errors)
df2['Date (Created)'] = pd.to_datetime(df2['Date (Created)'], errors='coerce')

# Define a cutoff date for "recent" leads (60 days ago)
recent_cutoff_date = pd.to_datetime('today') - pd.Timedelta(days=60)

# Filter for leads created within the last 60 days
recent_leads = df2[df2['Date (Created)'] >= recent_cutoff_date].copy()

# Define keywords/phrases indicating high priority in comments (expanded)
high_priority_keywords = ['meetup', 'interested', 'want to join', 'ready to sign', 'call back', 'schedule', 'demo', 'next steps']

# Function to check if any high-priority keyword is in the comment
def is_high_priority(comment):
    if isinstance(comment, str):
        comment_lower = comment.lower()
        for keyword in high_priority_keywords:
            if keyword in comment_lower:
                return True
    return False

# Identify high-priority recent leads based on comments
high_priority_recent_leads = recent_leads[recent_leads['Comment'].apply(is_high_priority)]

# Identify recently active leads (recent creation)
recently_active_leads = recent_leads[(recent_leads['Last Month'] > 0) | (recent_leads['This Week'] > 0) | (recent_leads['This Month'] > 0)]

print("High-Priority Recent Leads (Created in last 60 days with positive comments):")
if not high_priority_recent_leads.empty:
    print(high_priority_recent_leads)
else:
    print("No high-priority recent leads found based on the defined criteria.")

print("\nRecently Active Leads (Created in last 60 days with recorded activity):")
if not recently_active_leads.empty:
    print(recently_active_leads)
else:
    print("No recently active leads found based on the defined criteria.")
```

```python
import pandas as pd

# Assuming inactive_agents DataFrame is available

# Convert 'Date (Created)' to datetime objects (already done, but ensuring)
inactive_agents['Date (Created)'] = pd.to_datetime(inactive_agents['Date (Created)'], errors='coerce')

# Calculate Days Since Creation
current_date = pd.to_datetime('today')
inactive_agents['Days_Since_Creation'] = (current_date - inactive_agents['Date (Created)']).dt.days

# Extract Month and Quarter of Creation
inactive_agents['Creation_Month'] = inactive_agents['Date (Created)'].dt.month
inactive_agents['Creation_Quarter'] = inactive_agents['Date (Created)'].dt.quarter

# Binary Feature for Comment Presence
inactive_agents['Has_Comment'] = inactive_agents['Comment'].notna().astype(int)

# Keyword Features from Comments (Example - expand based on your data)
keywords = ['rates', 'whatsapp', 'meeting', 'no response', 'busy', 'left']
for keyword in keywords:
    inactive_agents[f'Comment_Has_{keyword.replace(" ", "_")}'] = inactive_agents['Comment'].str.lower().str.contains(keyword, na=False).astype(int)

print(inactive_agents[['Name', 'Date (Created)', 'Days_Since_Creation', 'Creation_Month', 'Creation_Quarter', 'Has_Comment'] + [col for col in inactive_agents.columns if 'Comment_Has_' in col]].head())

# Prepare data for clustering (excluding text columns)
clustering_data_184 = inactive_agents[[
    'Days_Since_Creation', 'Creation_Month', 'Creation_Quarter', 'Has_Comment'
] + [col for col in inactive_agents.columns if 'Comment_Has_' in col] + ['Owner', 'Lead Type']].copy()

# Convert categorical features to numerical using one-hot encoding
clustering_data_184 = pd.get_dummies(clustering_data_184, columns=['Owner', 'Lead Type'], drop_first=True)

print("\nPrepared data for clustering (first 5 rows):")
print(clustering_data_184.head())
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Scale the numerical features
scaler = StandardScaler()
numerical_features = ['Days_Since_Creation'] + [col for col in clustering_data_184.columns if 'Comment_Has_' in col]
scaled_numerical_data = scaler.fit_transform(clustering_data_184[numerical_features])

# Combine scaled numerical data with the rest of the (now numerical) data
clustering_data_scaled = pd.concat([pd.DataFrame(scaled_numerical_data, columns=numerical_features, index=clustering_data_184.index),
                                     clustering_data_184.drop(columns=numerical_features)], axis=1)

# Use the Elbow method to find the optimal number of clusters (k)
inertia = []
k_range = range(1, 11)  # Try 1 to 10 clusters
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(clustering_data_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow method
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Perform K-Means clustering with K=3
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
clusters = kmeans.fit_predict(clustering_data_scaled)

# Add the cluster labels back to the inactive_agents DataFrame
inactive_agents['Cluster'] = clusters

# Analyze the characteristics of each cluster
cluster_analysis = inactive_agents.groupby('Cluster').agg({
    'Days_Since_Creation': 'mean',
    'Creation_Month': lambda x: x.mode()[0] if not x.mode().empty else None,
    'Creation_Quarter': lambda x: x.mode()[0] if not x.mode().empty else None,
    'Has_Comment': 'mean',
    **{col: 'mean' for col in inactive_agents.columns if 'Comment_Has_' in col},
    'Owner': lambda x: x.mode()[0] if not x.mode().empty else None,
    'Lead Type': lambda x: x.mode()[0] if not x.mode().empty else None,
    'Name': 'count'  # Count of agents in each cluster
}).rename(columns={'Name': 'Cluster_Size'})

print("\nCluster Analysis (Mean/Mode Characteristics):")
print(cluster_analysis)
```
